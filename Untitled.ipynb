```python
import serial
import pandas as pd
import cv2
```

```python
ser = serial.Serial("COM3", 9600, timeout=1000) 
```

```python
flag = True
DataDict = {"AccX":[],"AccY":[],"AccZ":[],"GyroX":[],"GyroY":[],"GryoZ":[],"Time":[]}

  

while flag == True:    
    line = ser.readline()
    parts = line.decode().strip().split("/")
    
    print(parts)
    if len(parts) == 7:
        print(parts)
        DataDict["AccX"].append(parts[0])
        DataDict["AccY"].append(parts[1])
        DataDict["AccZ"].append(parts[2])
        DataDict["GyroX"].append(parts[3])
        DataDict["GyroY"].append(parts[4])
        DataDict["GryoZ"].append(parts[5])
        DataDict["Time"].append(parts[6])
    else:
        print("Ei hyva pomo")
        
    if cv2.waitKey(0) & 0xFF == ord('q'):
        flag = False
    
df = pd.DataFrame.from_dict(DataDict)
ser.close()
```
